This notebook is for me to have a way to practice resizing, saving, and manipulating images. The image that I'll be working on is called test_image.jpg and it's in the same directory as this notebook. It's 200 by 200 pixels. There is another called test_image_28_by_28 which is smaller.

```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

First I'll get the path to the image

```python
filename = "test_image.jpg"
this_files_directory = os.getcwd()
image_path = this_files_directory+"/"+filename
print("path to image is is: " + str(image_path))
```

Now let's take a look at the image. I'm opening it with np.asarray() because I'll be using that method later to enable myself to manipulate the shapes of image vectors. I want to make sure that it works well. I could just show the image by doing *img = Image.open(iamge_path)* and then running *plt.imshow(img)*

```python
image = np.asarray(Image.open(image_path))
plt.imshow(image)
```

Now I'll resize it to 28 by 28 pixels and view it again:

```python
image_to_resize = Image.open(image_path)
image_resized = image_to_resize.resize([28, 28])
plt.imshow(image_resized)
plt.show()
```

Now I'll reshape it into [784, 1], reshape it back, and view it again.

```python
image_to_reshape = np.asarray(image_resized)
np.shape(image_to_reshape)
```

Looks like I'll have to run my greyscale operation now to get rid of that third dimension: I'll make it greyscale the lazy way. (Check out https://e2eml.school/convert_rgb_to_grayscale.html for a better way to do this if it isn't good enough.) Then I'll view it again. I also want to log what the first pixel is:

```python
greyscale_image = np.mean(image_to_reshape, axis=2)
plt.imshow(image)
plt.show()
np.shape(greyscale_image)
greyscale_image[0][0]
```

There we go! Let's reshape that bad boy! Let's also check out the value of the first pixel:

```python
reshaped_image = greyscale_image.reshape([784, 1])
reshaped_again_image = reshaped_image.reshape([28, 28])
plt.imshow(reshaped_again_image)
plt.show()
reshaped_again_image[0][0]
```
